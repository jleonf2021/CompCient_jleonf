**<h1>Third task:
 Replace pattern in malformed ids (Cédula)</h1>**

```python
# Compiling the correct pattern for id
pattern = re.compile(r"\d{2}-\d{3}")
def replace_pattern(x):
  # Reviewing if the pattern of value match with wanted pattern
  match =  pattern.match(x)
  # If the pattern match return the same value
  if match:
    return x
  else:
    # If the pattern doesn't match replace - if exists
    x = x.replace("-","")
    # Select first two chars add - and finally add the rest of string 
    x = x[:2] + "-" + x[2:]
    return x
```

```python
def replace_pattern_2(x):
  # Reviewing if the pattern of value matches with wanted pattern
  match =  pattern.match(x)
  # If the pattern matches return the same value
  if match:
    return x
  else:
    # If the pattern doesn't match split string by -
    x = x.split("-")
    aux = ""
    # Order array in reverse 
    for i in reversed(x):
      aux = aux + i + "-"
    return aux[:-1]
```

```python
# Applying a method for replace mal formed ids in the column "Cedula"
homicides_df["CÉDULA"]= homicides_df["CÉDULA"].apply(replace_pattern_2)
```

```python
# Using regular expressions for validating if "CÉDULA" values match the pattern XX-XXX
cedula_malformed = homicides_df.loc[homicides_df["CÉDULA"].astype(str).apply(lambda x: (re.match("\d{2}-\d{3}", x) is None))]
```

**<h1>Fourth task:
 Remove invalid email addresses</h1>**

```python
# Verifying if the email has a valid pattern
def valid_email_pattern(x):
  # Split str by @
  x =  x.split("@")
  # if the email before @ has at least 4 chars it could be a valid email
  if(len(x[0])>3):
    # if the domain of email has .com or .edu.co is a valid email
    if(x[1].find(".com")!=-1 or x[1].find(".edu.co")!=-1):
      return True
  return False
```

```python
# Collecting malformed emails
email_malformed = homicides_df.loc[homicides_df["CORREO"].astype(str).apply(lambda x: (valid_email_pattern(x) == False))]
```

**<h1>Fifth task:
 Replace malformed days values with the most possible day</h1>**

```python
# Array with a valid days
valid_days = ["Lunes","Martes","Miercoles","Jueves", "Viernes","Sabado","Domingo"]
```

```python
# Checking what is the possible day for an entered value
def correct_days(x):
  aux = -1
  response = x
  if x not in valid_days:
    for j in valid_days:
      distance = pylev.levenshtein(x, j)
      if(distance < aux or aux == -1):
        aux = distance
        response = j
  return response
```

```python
# Test
correct_days("mercoles")
```

```python
# Correcting days in df
homicides_df["DIA"]= homicides_df["DIA"].apply(correct_days)
```

**<h1>Sixth task:
 Correct Born year, by year of record and age</h1>**

```python
# Function for calculate born year, by an input data
def correctBornYear(aux):
  # Convert date in str, split by - and finally select the first position (year)
  yearOfRecord = str(aux["FECHA"]).split("-")[0]
  # Get value of column EDAD
  age = aux["EDAD"]
  # Calculate born year by the mathematics operation year of record less age
  bornYear = int(yearOfRecord) - age
  return bornYear
```

```python
# Correct the values in column AÑO DE NACIMIENTO
homicides_df["AÑO DE NACIMIENTO"] = homicides_df.apply(lambda x: correctBornYear(x),axis=1)
```
